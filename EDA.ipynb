```python
# iqr
q1 = discount_coupon_df['Discount_pct'].quantile(0.25)
q3 = discount_coupon_df['Discount_pct'].quantile(0.75)

iqr = q3 - q1

upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr

print(q1, q3, upper_limit, lower_limit)

cond = (discount_coupon_df['Discount_pct'] > upper_limit) & (discount_coupon_df['Discount_pct'] < lower_limit)

discount_coupon_df[cond]
```

```python
discount_coupon_df = discount_coupon_df[~cond]
# discount_coupon_df = discount_coupon_df[~cond]
```

- 할인율
- 온라인 비용
- 오프라인 비용
- 주문수량
- 주문 건당 가격

```python
check_list = [marketing_spend_df['Offline_Spend'], marketing_spend_df['Online_Spend'], online_sales_df['Quantity'], online_sales_df['Avg_Price'], online_sales_df['Delivery_Charges']]
for i in range(5):
    # iqr
    q1 = check_list[i].quantile(0.25)
    q3 = check_list[i].quantile(0.75)

    iqr = q3 - q1

    upper_limit = q3 + 1.5*iqr
    lower_limit = q1 - 1.5*iqr
    
    cond = (check_list[i] > upper_limit) & (check_list[i] < lower_limit)

    print(f'{i+1}번 : {cond.sum()}')
    print(q1, q3, upper_limit, lower_limit)
```
